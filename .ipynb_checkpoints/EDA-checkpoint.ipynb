```python
(application_train['DAYS_BIRTH'] / -365).describe()
```

```python
# Описание столбца DAYS_EMPLOYED
application_train['DAYS_EMPLOYED'].describe()

# Построение гистограммы
application_train['DAYS_EMPLOYED'].plot.hist(title='Гистограмма дней занятости');
plt.xlabel('Дни занятости');
plt.show()
```

 в столбце DAYS_EMPLOYED существует аномалия: одно значение (365243) явно не соответствует реальным данным, так как оно указывает на занятость в течение 1000 лет. 

```python
# Флаг аномалий
anom = application_train[application_train['DAYS_EMPLOYED'] == 365243]
non_anom = application_train[application_train['DAYS_EMPLOYED'] != 365243]

# Доли дефолтов среди аномалий и неаномалий
print('Неаномалии имеют дефолт на %0.2f%% кредитов' % (100 * non_anom['TARGET'].mean()))
print('Аномалии имеют дефолт на %0.2f%% кредитов' % (100 * anom['TARGET'].mean()))
print('В данных %d записей с аномальными днями занятости' % len(anom))

# Обработка аномалий
application_train['DAYS_EMPLOYED_ANOM'] = application_train["DAYS_EMPLOYED"] == 365243
application_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace=True)

# Новая гистограмма после обработки
application_train['DAYS_EMPLOYED'].plot.hist(title='Гистограмма дней занятости после обработки');
plt.xlabel('Дни занятости');
plt.show()
```

Клиенты с аномальными значениями занятости имеют значительно более низкий уровень дефолтов. Это может быть важным сигналом для модели, так как эта группа отличается от общей массы.

Большинство категориальных переменных имеют относительно небольшое количество уникальных записей. Нам нужно будет найти способ работать с этими категориальными переменными!


```python
numerical_cols = X.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = X.select_dtypes(include=['object']).columns
```

```python
missing_categorical = X[categorical_cols].isnull().sum()
missing_numerical = X[numerical_cols].isnull().sum()

print("Missing values in categorical columns:")
print(missing_categorical.sort_values(ascending = False))

print("Missing values in numerical columns:")
print(missing_numerical.sort_values(ascending = False))
```

## Label Encoding and One-Hot Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Создаем объект LabelEncoder
le = LabelEncoder()
le_count = 0

for col in application_train:
    if application_train[col].dtype == 'object':
        # Если уникальных значений 2 или меньше, применяем Label Encoding
        if len(application_train[col].unique()) <= 2:
            # Тренируем LabelEncoder на тренировочных данных
            le.fit(application_train[col])
            
            # Трансформируем тренировочные и тестовые данные
            application_train[col] = le.transform(application_train[col])
            
            # Увеличиваем счетчик обработанных колонок
            le_count += 1

print(f'{le_count} колонок были закодированы с помощью Label Encoding.')

# Применяем One-Hot Encoding для остальных категориальных переменных
application_train = pd.get_dummies(application_train, drop_first=True) ## ДОДЕЛАТЬ, УДАЛЯЕТ НЕКОТОРЫЕ КОЛОНКИ


print('One-Hot Encoding завершен. Размеры данных:')
print(f'Тренировочные данные: {application_train.shape}')
```
